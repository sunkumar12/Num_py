```python
#increase size
plt.figure(figsize=(15,10))
a=np.array([1,2,3,4,5])
b=np.array([2,2,2,2,2])
c=np.array([3,2,4,6,5])

plt.plot(a, c='red',linestyle ='-.', label='population FY22')
plt.plot(c, c='green',linestyle ='--',label='population FY23')
plt.plot(b, linewidth=7., label='AVG population FY22')
plt.title("Project per month summary", fontsize=30,fontweight='bold')
plt.legend(loc='upper left',fontsize=10)
plt.xticks(range(0,len(a)),('JAN','FEB','MAR','APR','MAY'))
#plt.xticks([1,2,3,4,5,6])
plt.xlabel('Months')
plt.ylabel('Population')
plt.show()
```

```python
#Creating raw data
category = ['Account A','Account B','Account C','Account C']
Revenue = np.array([200,300,400,500])

# plotting it
plt.figure(figsize=(10,10))
plt.xticks(range(0,4),category)
plt.bar(range(0,4),Revenue)
plt.show()
```

```python
#Creating raw data
category = ['Account A','Account B','Account C','Account C']
Revenue = np.array([200,300,450,500])

# plotting it bar 
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.xticks(range(0,4),category)
plt.bar(range(0,4),Revenue)


# plotting it line 
plt.subplot(132)
plt.xticks(range(0,4),category)
plt.plot(range(0,4),Revenue, c='red')

# plotting it scatter
plt.subplot(133)
plt.xticks(range(0,4),category)
plt.scatter(range(0,4),Revenue, c='green')
plt.show()
```
